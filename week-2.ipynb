```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


print(is_prime(7))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

for i in range(3, 1000, 2):
    if is_prime(i) and is_prime(i+2):
        print((i, i+2))
```

Question 6

```python
marks = [20, 30, 35, 55, 70, 80, 90]

mean = sum(marks) / len(marks)


marks.sort()
n = len(marks)
if n % 2 == 0:
    median = (marks[n//2 - 1] + marks[n//2]) / 2
else:
    median = marks[n//2]

print("Mean:", mean)
print("Median:", median)
```

Question 7

```python
def decimal_to_binary(n):
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary


print(decimal_to_binary(10))
```

Question 8

```python
def prodDigits(n):
    product = 1
    while n > 0:
        product *= n % 10
        n //= 10
    return product


print(prodDigits(123))
```
