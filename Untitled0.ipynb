```python
def fib(n):
    a, b = 0, 1
    count = 0
    while count < n:
        print(a, end=" ")
        a, b = b, a + b
        count += 1


n = int(input("Enter number of number: "))
fib(n)
```

```python
def myPow(x, n):
    result = 1
    count = 0
    while count < n:
        result *= x
        count += 1
    return result

print(myPow(2, 3))
```

```python
def SumDigits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total += digit
        n //= 10
    print("Sum of digits:", total)


SumDigits(234)
```

```python
def squaredSum(arr):
    total = 0
    for num in arr:
        total += num * num
    return total


print(squaredSum([2, 3, 4]))
```

```python
def sumDigits(num):
    total = 0
    while num > 0:
        total += num % 10
        num //= 10
    return total


print(sumDigits(666))
```
