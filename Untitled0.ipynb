```python
def get_unique_elements(input_list):
    # Converting to a set removes duplicates, then we convert back to a list
    return list(set(input_list))

# Testing the function
sample_list = [1, 2, 3, 3, 3, 3, 4, 5]
print(f"Unique List: {get_unique_elements(sample_list)}")
```
